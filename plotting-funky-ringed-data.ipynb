```python
def comparing_LCs(parent, LCs, labels, letters, tr_dur, fis, fes, spangs):
    starts = (LCU.get_transit_start(LCs, 1800)*2).astype(int)
    lengths = []
    for i in range(len(labels)):
        lengths.append(np.argmax(LCs[i][starts[i]:] == 1))
    
    max_length = max(lengths)
        
    centered_LCs = LCs[:, :(max_length+4)].copy()
    for i in range(len(labels)):
        print(i)
        centered_LCs[i] = LCs[i][int(starts[i]-2):int(parent_CD_starts[i]+(max_length+2))]

    max_depth = np.min(centered_LCs)
    x = np.linspace(0, (max_length+3)/2, (max_length+4))-0.5

    fig, axes = pl.subplots(3, 1)
    axes[0].plot(np.linspace(0, 144, 145)/2, LCs[0], label="Dataset: Non-Ringed - Realization: "+labels[0])
    axes[1].plot(x, centered_LCs[0])
    axes[2].plot(x, centered_LCs[0])
    
    for i in range(len(labels)-1):
        axes[0].plot(np.linspace(0, 144, 145)/2, LCs[i+1],
                     label='Dataset: '+letters[i]+' - Realization: '+labels[i+1]+'\nInt= '+str(fis[i])[:4]+', Ext= '+str(fes[i])[:4]+', spang= '+str(spangs[i])+'\n')
        axes[1].plot(x, centered_LCs[i+1])
        axes[2].plot(x, centered_LCs[i+1])
    
    axes[0].set_title('How each simulated LC looks like:')
    axes[0].set_xlabel('Hours', loc='right')
    axes[0].set_ylabel('Normalized Flux')
    axes[1].set_title('Comparing transit shapes (Zoom X):')
    axes[1].set_xlabel('Hours since start of transit', loc='right')
    axes[1].set_ylabel('Normalized Flux')
    axes[2].set_title('Comparing transit shapes (Zoom Y):')
    axes[2].set_xlabel('Hours since start of transit', loc='right')
    axes[2].set_ylabel('Normalized Flux')
    axes[2].set_ylim(max_depth + (1-max_depth)*0.90, 1+max_depth*0.0001)
    
    axes[0].legend(bbox_to_anchor=(1, 1), fontsize='6')
    
    pl.subplots_adjust(hspace=0.36)
    fig.suptitle('Parent System: '+parent+' (Tran. dur.: ~'+str(tr_dur)+' hrs.)', fontsize=16)
    
    fig.set_size_inches(7, 10)
    
    return fig
    
    
    
```
