```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as pl
import LCUtils as LCU
%matplotlib inline
```

```python
A_curves = np.loadtxt('Datasets/Funky-rings/A-ringed-curves.txt')
B_curves = np.loadtxt('Datasets/Funky-rings/B-ringed-curves.txt')
C_curves = np.loadtxt('Datasets/Funky-rings/C-ringed-curves.txt')
D_curves = np.loadtxt('Datasets/Funky-rings/D-ringed-curves.txt')
E_curves = np.loadtxt('Datasets/Funky-rings/E-ringed-curves.txt')
F_curves = np.loadtxt('Datasets/Funky-rings/F-ringed-curves.txt')
```

```python
A_cropped_curves = LCU.LC_cropper(A_curves, 145, 1800, False)[0]
B_cropped_curves = LCU.LC_cropper(B_curves, 145, 1800, False)[0]
C_cropped_curves = LCU.LC_cropper(C_curves, 145, 1800, False)[0]
D_cropped_curves = LCU.LC_cropper(D_curves, 145, 1800, False)[0]
E_cropped_curves = LCU.LC_cropper(E_curves, 145, 1800, False)[0]
F_cropped_curves = LCU.LC_cropper(F_curves, 145, 1800, False)[0]
```

```python
A_start = LCU.get_transit_start(A_cropped_curves, 1800)
B_start = LCU.get_transit_start(B_cropped_curves, 1800)
C_start = LCU.get_transit_start(C_cropped_curves, 1800)
D_start = LCU.get_transit_start(D_cropped_curves, 1800)
E_start = LCU.get_transit_start(E_cropped_curves, 1800)
F_start = LCU.get_transit_start(F_cropped_curves, 1800)

```

```python
fig, ax = pl.subplots(2, 3)
ax[0, 0].hist(A_start, density=True, bins=145)
ax[0, 1].hist(B_start, density=True, bins=145)
ax[0, 2].hist(C_start, density=True, bins=145)
ax[1, 0].hist(D_start, density=True, bins=145)
ax[1, 1].hist(E_start, density=True, bins=145)
ax[1, 2].hist(F_start, density=True, bins=145)

ax[0, 0].set_title('Funky Rings: A')
ax[0, 1].set_title('Funky Rings: B')
ax[0, 2].set_title('Funky Rings: C')
ax[1, 0].set_title('Funky Rings: D')
ax[1, 1].set_title('Funky Rings: E')
ax[1, 2].set_title('Funky Rings: F')

ax[0, 0].set_ylabel('Relative Freq.')
ax[1, 0].set_ylabel('Relative Freq.')
ax[1, 0].set_xlabel('Hours')
ax[1, 1].set_xlabel('Hours')
ax[1, 2].set_xlabel('Hours')

pl.subplots_adjust(hspace=0.36)
fig.suptitle('Time at which transit starts', fontsize=15)
fig.set_size_inches((10, 5))
```
