```python
ringed_parameters = pd.read_csv('Datasets/ringed-parameters.csv', index_col='Unnamed: 0')
ringed_parameters
```

```python
dataframes = []

number_batch = 54
size_batch = 10


for i in range(0, number_batch):
    if i!=53:
        dataframes.append(ringed_parameters[i*size_batch:i*size_batch+size_batch])
    if i==53:
        dataframes.append(ringed_parameters[i*size_batch:])
        
def simulator(x):
    return LCG.LC_ringed_simulator(x, 'ID', 'pl_radj', 'st_mass', 'st_rad', 'pl_orbsmax', 'pl_orbeccen',
                                           'pl_orbincl', 'pl_orblper', 'ring_ext_plrad', 'ring_int_plrad', 'ring_incl_deg', 2, 1800)
    
if __name__ == "__main__":
    with Pool() as pool:
        LC = pool.map(simulator, dataframes)
```

```python
LC_array_almost = np.array(LC[:-1], dtype=np.object_)
LC_array_almost = LC_array_almost.transpose(1, 0, 2).reshape(3, 530)
LC_array_last = np.array(LC[-1], dtype=np.object_)
LC_array_last = LC_array_last.reshape(3, 4)
whole_LC = np.hstack([LC_array_almost, LC_array_last])
only_LCS = np.array(whole_LC[2])
LC_list = []
for i in range(only_LCS.shape[0]):
    LC_list.append(only_LCS[i])

only_LCS = np.array(LC_list)
```

```python
np.savetxt('Datasets/ringed-ids.txt', whole_LC[0, :].astype(str), fmt='%s')
np.savetxt('Datasets/ringed-periods.txt', whole_LC[1, :].astype(float))
np.savetxt('Datasets/ringed-curves.txt', only_LCS)
```
