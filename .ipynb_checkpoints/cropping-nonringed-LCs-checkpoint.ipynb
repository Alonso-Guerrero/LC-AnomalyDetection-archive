```python
import numpy as np
import pandas as pd
import LCUtils as LCU
```

```python
LCs_1 = []
transit_durations_1 = np.array([])
transit_exists_1 = np.array([])
```

```python
for i in range(80):
    lcs = np.loadtxt('Datasets/curves_1_'+str(i)+'.txt', float)
    print('Starting with file:', i)
    results = LCU.LC_cropper(lcs, 145, 1800, True)
    LCs_1.append(results[0])
    transit_durations_1 = np.hstack((transit_durations_1, results[1]))
    transit_exists_1 = np.hstack((transit_exists_1, results[2]))
    print('')
```

```python
LCs_1_array = np.vstack(LCs_1)
```

```python
np.savetxt('Datasets/non-ringed-cropped-curves-1.txt', LCs_1_array)
np.savetxt('Datasets/non-ringed-transit-durations-1.txt', transit_durations_1)
np.savetxt('Datasets/non-ringed-transit-masks-1.txt', transit_exists_1)
```

```python
LCs_2 = []
transit_durations_2 = np.array([])
transit_exists_2 = np.array([])
```

```python
for i in range(80):
    lcs = np.loadtxt('Datasets/curves_2_'+str(i)+'.txt', float)
    print('Starting with file:', i)
    results = LCU.LC_cropper(lcs, 145, 1800, True)
    LCs_2.append(results[0])
    transit_durations_2 = np.hstack((transit_durations_2, results[1]))
    transit_exists_2 = np.hstack((transit_exists_2, results[2]))
    print('')
```

```python
LCs_2_array = np.vstack(LCs_2)
```

```python
np.savetxt('Datasets/non-ringed-cropped-curves-2.txt', LCs_2_array)
np.savetxt('Datasets/non-ringed-transit-durations-2.txt', transit_durations_2)
np.savetxt('Datasets/non-ringed-transit-masks-2.txt', transit_exists_2)
```

```python
all_LCs = np.vstack([LCs_1_array, LCs_2_array])
all_transit_durations = np.hstack((transit_durations_1, transit_durations_2))
all_transit_exists = np.hstack((transit_exists_1, transit_exists_2))
```

```python
np.savetxt('Datasets/non-ringed-cropped-curves.txt', all_LCs)
np.savetxt('Datasets/non-ringed-transit-durations.txt', all_transit_durations)
np.savetxt('Datasets/non-ringed-transit-masks.txt', all_transit_exists)
```

```python
##Now, appending the transit durations to the synthetic parameter grid:
df = pd.read_csv('Datasets/non-ringed-parameters.csv', index_col='Unnamed: 0')
df = pd.concat([df, pd.DataFrame(all_transit_exists, columns=['transit_exists']), pd.DataFrame(all_transit_durations, columns=['pl_trandur'])], axis=1)
```

```python
df.to_csv('Datasets/non-ringed-parameters.csv')
```
