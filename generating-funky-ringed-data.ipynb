```python
def produce_LC(dataset, number_batch, size_batch, n_spangles):
    dataframes = []
    for i in range(0, number_batch):
        if i!=(number_batch-1):
            dataframes.append(dataset[i*size_batch:i*size_batch+size_batch])
        if i==(number_batch-1):
            dataframes.append(dataset[i*size_batch:])
      
    def simulator(x):
        return LCG.LC_ringed_simulator(x, 'ID', 'pl_radj', 'st_mass', 'st_rad', 'pl_orbsmax', 'pl_orbeccen',
                                           'pl_orbincl', 'pl_orblper', 'ring_ext_plrad', 'ring_int_plrad', 'ring_incl_deg', n_spangles, 2, 1800)
    
    if __name__ == "__main__":
        with Pool() as pool:
            LC = pool.map(simulator, dataframes)
        
    return LC
    
    
```

```python
def reshape_results(results, almost, last):
    LC_array_almost = np.array(results[:-1], dtype=np.object_)
    LC_array_almost = LC_array_almost.transpose(1, 0, 2).reshape(3, almost)
    LC_array_last = np.array(results[-1], dtype=np.object_)
    LC_array_last = LC_array_last.reshape(3, last)
    whole_LC = np.hstack([LC_array_almost, LC_array_last])
    only_LCS = np.array(whole_LC[2])
    LC_list = []
    for i in range(only_LCS.shape[0]):
        LC_list.append(only_LCS[i])

    only_LCS = np.array(LC_list)
    
    return [whole_LC[0, :], whole_LC[1, :], only_LCS]
```

```python
def concatenate_dataframe(dataset, periods, masks, durations):
    periods_df = pd.DataFrame(periods, columns=['pl_orbper_secs'])
    masks_df = pd.DataFrame(masks, columns=['transit_exists'])
    durations_df = pd.DataFrame(durations, columns=['pl_trandur'])
    
    return pd.concat([dataset, periods_df, masks_df, durations_df], axis=1)
```
