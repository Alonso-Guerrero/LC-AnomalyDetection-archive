```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as pl
%matplotlib inline
```

```python
parent_parameters= pd.read_csv('Datasets/parent-parameters.csv', index_col='Unnamed: 0')
parent_ids = parent_parameters['ID'].values.astype(str)
```

```python
A_ids = np.loadtxt('Datasets/Funky-rings/A-ringed-ids.txt', str)
B_ids = np.loadtxt('Datasets/Funky-rings/B-ringed-ids.txt', str)
C_ids = np.loadtxt('Datasets/Funky-rings/C-ringed-ids.txt', str)
D_ids = np.loadtxt('Datasets/Funky-rings/D-ringed-ids.txt', str)
E_ids = np.loadtxt('Datasets/Funky-rings/E-ringed-ids.txt', str)
F_ids = np.loadtxt('Datasets/Funky-rings/F-ringed-ids.txt', str)
```

```python
A_ids = np.array([s[:2] for s in A_ids])
B_ids = np.array([s[:2] for s in B_ids])
C_ids = np.array([s[:2] for s in C_ids])
D_ids = np.array([s[:2] for s in D_ids])
E_ids = np.array([s[:2] for s in E_ids])
F_ids = np.array([s[:2] for s in F_ids])
```

```python
A_hist, _ = np.histogram(A_ids, bins=np.append(parent_ids, parent_ids[-1]+'DV'))
B_hist, _ = np.histogram(B_ids, bins=np.append(parent_ids, parent_ids[-1]+'DV'))
C_hist, _ = np.histogram(C_ids, bins=np.append(parent_ids, parent_ids[-1]+'DV'))
D_hist, _ = np.histogram(D_ids, bins=np.append(parent_ids, parent_ids[-1]+'DV'))
E_hist, _ = np.histogram(E_ids, bins=np.append(parent_ids, parent_ids[-1]+'DV'))
F_hist, _ = np.histogram(F_ids, bins=np.append(parent_ids, parent_ids[-1]+'DV'))

```

```python
fig, ax = pl.subplots(1, 1)
ax.barh(parent_ids, A_hist)
ax.set_title('Funky Rings: A')
ax.set_xlabel('Freq.')
ax.set_ylabel('Parent System')
fig.set_size_inches((5, 16))
```

```python
fig, ax = pl.subplots(1, 1)
ax.barh(parent_ids, B_hist)
ax.set_title('Funky Rings: B')
ax.set_xlabel('Freq.')
ax.set_ylabel('Parent System')
fig.set_size_inches((5, 16))
```

```python
fig, ax = pl.subplots(1, 1)
ax.barh(parent_ids, C_hist)
ax.set_title('Funky Rings: C')
ax.set_xlabel('Freq.')
ax.set_ylabel('Parent System')
fig.set_size_inches((5, 16))
```

```python
fig, ax = pl.subplots(1, 1)
ax.barh(parent_ids, D_hist)
ax.set_title('Funky Rings: D')
ax.set_xlabel('Freq.')
ax.set_ylabel('Parent System')
fig.set_size_inches((5, 16))
```

```python
fig, ax = pl.subplots(1, 1)
ax.barh(parent_ids, E_hist)
ax.set_title('Funky Rings: E')
ax.set_xlabel('Freq.')
ax.set_ylabel('Parent System')
fig.set_size_inches((5, 16))
```

```python
fig, ax = pl.subplots(1, 1)
ax.barh(parent_ids, F_hist)
ax.set_title('Funky Rings: F')
ax.set_xlabel('Freq.')
ax.set_ylabel('Parent System')
fig.set_size_inches((5, 16))
```
